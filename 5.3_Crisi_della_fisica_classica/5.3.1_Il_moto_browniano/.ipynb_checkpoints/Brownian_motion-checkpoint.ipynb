#Il moto browniano
##Un po' di storia
Già nella seconda metà dell’Ottocento,
grazie soprattutto al lavoro teorico di
James Clerk Maxwell (1831-1879) e di
Ludwig Boltzmann (1844-1906), si era
riusciti a giustificare il comportamento di
un gas ammettendo che esso fosse forma-
to da un numero enorme di «grani» indi-
pendenti (chiamati molecole), dotati di
un moto continuo, velocissimo e casuale.

Con il semplice ausilio di un microscopio ottico la situazione osservata
cambia drasticamente: una piccola particella immersa in acqua è tutt'altro
che ferma, animata da un movimento perfettamente disordinato. Tale fenomeno viene denominato moto browniano, in onore del botanico inglese Robert
Brown, che lo osservò chiaramente per la prima volta nel 1827.
A rendere possibile l'osservazione fu l'introduzione in quel periodo dei
primi obiettivi acromatici. Brown stesso cita dei precursori nella scoperta
del fenomeno, ma questi ( Buffon e Spallazzi), probabilmente a causa delle diffcoltà tecniche, avevano visto nelle particelle degli "*animaletti microscopici danzanti*".

#Proprietà
1. Il fenomeno si presenta in tutti i fluidi
1. con una attività inversamente proporzionale alla viscosità del fluido in questione
1. Non è riconducibile a correnti di convezione termica o alle vibrazioni del supporto del recipiente
1. il moto non si arresta mai è "eterno e spontaneo"
1. Il moto è a zig-zag con bruschi cambi di direzione


```python

n = 5000
x = np.cumsum(np.random.randn(n))
y = np.cumsum(np.random.randn(n))
# We add 10 intermediary points between two
# successive points. We interpolate x and y.
k = 10
x2 = np.interp(np.arange(n * k), np.arange(n) * k, x)
y2 = np.interp(np.arange(n * k), np.arange(n) * k, y)
fig, ax = plt.subplots(1, 1, figsize=(8, 8))
# Now, we draw our points with a gradient of colors.
ax.scatter(x2, y2, c=range(n * k), linewidths=0,
           marker='o', s=3, cmap=plt.cm.jet,)
ax.axis('equal')
ax.set_axis_off()
```

#La teoria di Einstein

La teoria di Einstein si sviluppa a partire dalla sua tesi di dottorato, discussa
il 30 Aprile 1905, in cui presenta un metodo per la determinazione della
dimensione molecolari e del numero di Avogadro dal coefficiente di viscosità
di soluzioni zuccherine.
Undici giorni dopo la rivista Annalen der Physik pubblica un articolo
intitolato: *Die von der Molekularkinetischen Theorie der Wärme gefordete Bewegung von in ruhenden Flüssigkeiten suspendierten Teilchen* , in cui
Einstein estende i ragionamenti fatti per le molecole di zucchero al caso di
particelle microscopiche. Di fatto propone una teoria quantitativa del moto
browniano.

##Il modello diEinstein
Einstein immagina ad un gas di sfere rigide in di massa piccola in cui è inserita una sfera di dimensioni maggiori e di massa ben più grande. Inoltre semplifica il problema ragionando in una sola dimensione: $x$.
###Legge di Stokes
Einstein parte dall'equazione di Stokes che descrive la resistenza viscosa di un fluido:
$$
F_v=6\pi\eta a v
$$
Dove:
* $\eta$ rappresenta la viscosità del fluido
* $a$ il diametro della sfera che si muove nel fluido
* $v$ la velocità

Immagina inoltre che il *salto* della particella debba essere deterrminato da una differenza di pressione, una sorta di pressione osmotica, determinata da un **gradiente di concentrazione** del liquido, che seppur minimo permette di dare alla particella un impulso in una direzione. Indicando con $F_v$ la vorza agente sulle sferette ideali, dovuta la gradiente di concentrazione, Einstein arriva alla relazione
$$
f_v\frac{\rho N_A}{m}=\frac{RT}{m}\frac{d\rho}{dx}
$$
L'intuizione di Einstein consiste nell'eguagliare la pressione osmotica con
resistenza del solvente al movimento delle sferette , in condizioni di equilibrio
dinamico. Così facendo si potrebbe già arrivare ad una relazione tra grandezze
note o misurabili sperimentalmente (uguagliando le due formule precedenti):
$$
6\pi\eta a v\rho = \frac{RT}{N_A}\frac{d\rho}{dx}
$$
* $\frac{d\rho}{dx}$ è la concentrazione
* $R$ è la costante dei gas ideali
* $N_A$ è il numerod il Avogadro

Ricaviamo la velocità e moltiplicandola per la concentrazione si ottiene  **il numero di particelle che attraversano l'area unitaria per unità di
tempo** (per semplicitò in una dimensione). Cioè:
$$
v \rho= \frac{1}{6\pi\eta a }\frac{RT}{ N_A}\frac{d\rho}{dx}
$$

Ma all'equilibrio questo è anche uguale al comvenzionale concetto di **diffusione** dove
$$
D \frac{d\rho}{dx} =\frac{\text{particelle}}{cm^2s} 
$$

combinando le due equazioni si ricava:
$$
v \rho=D \frac{d\rho}{dx} \Rightarrow D= \frac{1}{6\pi\eta a }\frac{RT}{ N_A}
$$
